```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.style as style
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme()
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    col = "time", #facet
    hue = "smoker", #color
    style = "smoker", #shape
    size = "size"
)
```

```python
sns.relplot(
    data = tips)
```

```python
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    col = "time"
)
```

```python
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    col = "time",
    hue = "smoker", #color
    style = "smoker", #shape
    size = "size",
    kind = "line" #create line graph
)
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.histplot(
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    multiple = "stack" #stack bars
)
```

```python
sns.kdeplot( #kernel density estimate plot
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    multiple = "stack" #stack bars
)
```

```python
sns.displot(
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    multiple = "stack" #stack bars
)
```

```python
sns.histplot(
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    multiple = "stack", #stack bars
    col = "species"
)
```

```python
sns.displot(
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    multiple = "stack", #stack bars
    col = "species"
)
```

```python
sns.displot(
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    multiple = "stack", #stack bars
    col = "species",
    kind = "kde"
)
```

```python
sns.displot(
    data = penguins,
    x = "flipper_length_mm",
    hue = "species",
    col = "species",
    kind = "ecdf" #cumulative density plot
)
```

```python
sns.jointplot(
    data = penguins,
    x = "flipper_length_mm",
    y = "bill_length_mm",
    hue = "species",
    alpha = 0.5 #transparency
)
```

```python
sns.pairplot( #scatterplot matrix
    data = penguins,
    hue = "species"
)
```

```python
sns.set_style("whitegrid") #theme
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)
```
